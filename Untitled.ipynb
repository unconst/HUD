```python
def _trans(block):
    sub = bittensor.subtensor(network = 'nakamoto' )
    block_hash = sub.substrate.get_block_hash(block)
    result = sub.substrate.get_block(block_hash=block_hash)
    print("block_number", block_number, "block_hash", block_hash, "extrinsics", len(result['extrinsics']) )
    for r in result['extrinsics']:
        if r.value["call"]["call_module"] == "Balances":
            vals = {
                'hash': r.value['extrinsic_hash'],
                'block': block,
                'source': r.value['address'],
                'dest': r.value["call"]["call_args"][0]['value']
            }
            with open("transfers.txt", "a") as transfers:
                transfers.write( str(vals) + "\n")
    return True
    
```
